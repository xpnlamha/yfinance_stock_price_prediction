```python
df = pd.read_csv('../../Dataset/MSFT.csv')
```

```python
# Chia train test validate theo tỉ lệ 8:2
df=df['Close']

df = df.dropna() # Drop missing values
df = df.reset_index(drop=True) # Reset the index

# Split the data into training, testing, and validation sets
train_size = int(0.8 * len(df))
test_size = int(0.2 * len(df))

train_data = df[:train_size]
test_data = df[train_size:]
```

```python
# # Chia train test validate theo tỉ lệ 9:1
# df=df['Close']

# df = df.dropna() # Drop missing values
# df = df.reset_index(drop=True) # Reset the index

# # Split the data into training, testing, and validation sets
# train_size = int(0.9 * len(df))
# test_size = int(0.1 * len(df))

# train_data = df[:train_size]
# test_data = df[train_size:]
```

# 3. Quá trình Training

```python
x_train = np.array(train_data.index).reshape(-1, 1)
y_train = np.array(train_data)

# Train the linear regression model
model = LinearRegression()
model.fit(x_train, y_train)
```

# 4. Quá trình testing

```python

x_test = np.array(test_data.index).reshape(-1, 1)
y_test = np.array(test_data)
y_pred = model.predict(x_test)
```

# 6. Quá trình tạo index predict 30, 60, 90 ngày tiếp theo

```python
last_index =  df.index[-1]
```

```python
last_data_30_days = pd.RangeIndex(start=last_index, stop=last_index+30, step=1)

# Create an array of 30 consecutive integers starting from last_index
x_next_30_days = np.array(range(last_index+1, last_index+31)).reshape(-1, 1)

# Predict the closing prices for the next 30 days
y_next_30_days = model.predict(x_next_30_days)

# Print the predicted closing prices for the next 30 days
print('Predicted closing prices for the next 30 days:')
print(y_next_30_days)
```

```python
last_data_60_days = pd.RangeIndex(start=last_index, stop=last_index+60, step=1)

# Create an array of 30 consecutive integers starting from last_index
x_next_60_days = np.array(range(last_index+1, last_index+61)).reshape(-1, 1)

# Predict the closing prices for the next 30 days
y_next_60_days = model.predict(x_next_60_days)

# Print the predicted closing prices for the next 30 days
print('Predicted closing prices for the next 60 days:')
print(y_next_60_days)
```

```python
last_data_90_days = pd.RangeIndex(start=last_index, stop=last_index+90, step=1)

# Create an array of 30 consecutive integers starting from last_index
x_next_90_days = np.array(range(last_index+1, last_index+91)).reshape(-1, 1)

# Predict the closing prices for the next 30 days
y_next_90_days = model.predict(x_next_90_days)

# Print the predicted closing prices for the next 30 days
print('Predicted closing prices for the next 90 days:')
print(y_next_90_days)
```

# 7. Đánh giá độ chính xác thuật toán

```python
#RMSE
test_rmse = np.sqrt(np.mean((y_pred - y_test)**2))
print('Testing RMSE:', test_rmse)

#MAPE
test_mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print('Testing MAPE:', test_mape)

#MAE
test_mae = np.mean(np.abs(y_pred - y_test))
print('Testing MAE:', test_mae)
```

# 8.Vẽ hình

```python
plt.figure(figsize=(16,10))
plt.plot(train_data.index, train_data)
plt.plot(test_data.index, test_data)
plt.plot(test_data.index, y_pred)
plt.plot(last_data_30_days,y_next_30_days)
plt.plot(last_data_60_days,y_next_60_days)
plt.plot(last_data_90_days,y_next_90_days)
plt.legend(['Train', 'Test', 'Predictions','Next30Day','Next60Day','Next90Day'])
plt.show()
```
